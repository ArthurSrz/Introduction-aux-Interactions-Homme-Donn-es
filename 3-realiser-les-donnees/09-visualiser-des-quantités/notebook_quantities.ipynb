```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
birds = pd.read_csv("data/birds.csv")
birds.head()
```

```python
wingspan = birds["MaxWingspan"]
wingspan.plot()
```

```python
plt.title('Max Wingpsan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.xticks(rotation=45)
x = birds["Name"]
y = birds["MaxWingspan"]

plt.plot(x, y)
plt.show()
```

```python
plt.title('Max Wingpsan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

for i in range(len(birds)):
    x = birds["Name"][i]
    y = birds["MaxWingspan"][i]
    plt.plot(x, y, 'bo')
    if birds["MaxWingspan"][i] > 500:
        plt.text(x, y * (1-0.05), birds["Name"][i], fontsize=8)

plt.show()
```

```python
plt.title("Max Wingspan in Centimeters")
plt.ylabel("Wingspan (CM)")
plt.xlabel("Birds")
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
for i in range(len(birds)):
    x = birds["Name"][i]
    y = birds["MaxWingspan"][i]
    if birds["Name"][i] not in ["Bald eagle","Prairie falcon"]:
        plt.plot(x, y, 'bo')

plt.show()
        
        
```

```python
birds.plot(x="Category",
           kind="bar",
           stacked=True,
           title = "Birds by Category",)
```

```python
category_count = birds.value_counts(birds["Category"].values, sort=True)
maxlength = birds['MaxLength']
#sort maxlength
maxlength = maxlength.sort_values(ascending=False)
```

```python
plt.rcParams["figure.figsize"] = (6, 12)
```

```python
category_count.plot.barh()
```

```python
plt.barh(y=birds["Category"], width=maxlength)
```

```python
minLength = birds["MinLength"]
minLength = minLength.sort_values(ascending = False)
maxLength = birds["MaxLength"]
maxLength = maxLength.sort_values(ascending = False)
category = birds['Category']

plt.barh(category, maxLength, label="Max Length")
plt.barh(category, minLength, label="Min Length")

plt.show()
```
